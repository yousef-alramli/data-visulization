# analyzing data


## importing libraries

```python
import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd 
```

## Set default plotter to be seaborn

```python
sns.set()
```

## getting the data from sns data sets

```python
df = sns.load_dataset("iris")
data = pd.DataFrame(data=df)
```

## drawing figure for analysis

```python
sns.set_theme(style="whitegrid")

sns.relplot(x="sepal_length", y="sepal_width", hue="species", alpha=.5, palette="muted",height=6, data=data)
```

## summary

- As you can see in the figure the setota species have more sepal width in general than the others but also the sepal length is les than the other species

- for the versicolor and virginica they look similar but the virginica  have more sepals are higher at length
