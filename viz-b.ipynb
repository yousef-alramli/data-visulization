# analyzing data


## importing libraries

```python
import matplotlib.pyplot as plt

import seaborn as sns

import pandas as pd

```

## getting the data from sns data sets

```python
sns.set()
sns.get_dataset_names()
```

```python
df = sns.load_dataset("mpg")
data = pd.DataFrame(data=df)
```

```python
sns.relplot(x="horsepower", y="mpg", hue="cylinders", size="model_year",sizes=(25, 250), alpha=.5, palette="muted",height=6, data=data)
```

## Summary
- As we can see in the figure as more the selenders are the less mpg there is 

- the same for horse power it have Inverse relationship with the mpg

- At last the model year, it seems that the modren cars have more mpg
